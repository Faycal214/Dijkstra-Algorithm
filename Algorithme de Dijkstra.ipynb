# Zlgorithme de Dijkstra

## Description
Le problème de cheminement, également connu sous le nom de problème du plus court chemin, est un problème classique en informatique et en mathématiques discrètes. Il consiste à trouver le chemin le plus court ou le plus optimal entre deux points dans un graphe pondéré, où chaque arête a un poids associé. Ce problème est fondamental dans de nombreux domaines et trouve des applications pratiques dans les réseaux de communication, la planification de trajets, la logistique, la cartographie, la robotique, etc.

## Applications
- **Réseaux de communication** : Routage de paquets dans les réseaux informatiques, routage de données sur Internet, etc.
- **Planification de trajets** : Calcul d'itinéraires optimaux pour la navigation automobile, les applications de covoiturage, les applications de transport en commun, etc.
- **Logistique** : Optimisation des chaînes d'approvisionnement, de la distribution de marchandises, etc.
- **Cartographie** : Calcul d'itinéraires pour les applications de cartographie en ligne, les systèmes de navigation GPS, etc.
- **Robotique** : Planification de mouvement pour les robots autonomes, les drones, etc.

## Algorithmes
Plusieurs algorithmes sont utilisés pour résoudre le problème de cheminement :
- **Algorithme de Dijkstra** : Utilisé pour trouver le plus court chemin entre un sommet source et tous les autres sommets dans un graphe pondéré à arêtes non négatives.
- **Algorithme de Bellman-Ford** : Utilisé pour trouver le plus court chemin entre un sommet source et tous les autres sommets dans un graphe pondéré, même en présence d'arêtes négatives.
- **Algorithme de Floyd-Warshall** : Utilisé pour trouver les plus courts chemins entre tous les paires de sommets dans un graphe pondéré, même en présence d'arêtes négatives.

## Algorithme de Dijkstra
L'algorithme de Dijkstra est l'un des algorithmes les plus populaires pour résoudre le problème de cheminement. Il fonctionne en explorant sélectivement les nœuds du graphe à partir du nœud source, en choisissant à chaque étape le nœud le plus proche de la source jusqu'à ce que le nœud destination soit atteint. L'algorithme maintient une liste des distances les plus courtes connues entre le nœud source et chaque nœud du graphe, et il utilise une structure de données telle qu'une file de priorité pour sélectionner efficacement le prochain nœud à explorer.

## Pourquoi utiliser l'algorithme de Dijkstra ?

Dans le contexte du problème de cheminement, l'algorithme de Dijkstra est un choix approprié pour les raisons suivantes :

1. **Arêtes non négatives** : Dijkstra fonctionne efficacement dans les graphes où toutes les arêtes ont des poids non négatifs. Cela en fait une option idéale pour les problèmes où les poids des arêtes représentent des distances, des temps de trajet ou des coûts, qui sont naturellement non négatifs.

2. **Efficacité** : L'algorithme de Dijkstra est relativement efficace pour résoudre des problèmes de cheminement dans des graphes de taille modérée à grande. Sa complexité temporelle dépend du nombre de nœuds et d'arêtes dans le graphe, mais il est souvent plus rapide que d'autres algorithmes pour les graphes avec des arêtes non négatives.

3. **Fiabilité** : Dijkstra garantit de trouver le chemin le plus court entre un nœud source et tous les autres nœuds dans le graphe, à condition que toutes les arêtes soient non négatives. Cela en fait un choix fiable pour de nombreuses applications où la précision et la fiabilité sont importantes.

4. **Facilité d'implémentation** : L'algorithme de Dijkstra est relativement simple à comprendre et à implémenter. Il repose sur des principes de base comme la recherche du plus court chemin dans un graphe en utilisant des structures de données comme les files de priorité. Cela le rend accessible même aux débutants en algorithmique.

En résumé, l'algorithme de Dijkstra est souvent choisi pour sa compatibilité avec les graphes pondérés à arêtes non négatives, son efficacité, sa fiabilité et sa facilité d'implémentation. Il est particulièrement adapté aux applications où les distances ou les coûts entre les nœuds sont des valeurs positives et où l'on a besoin de trouver le chemin le plus court entre un nœud source et tous les autres nœuds dans le graphe.

# Algorithme de Dijkstra

Prérequis : Les longueurs sont positives ou nulles.
Principe : Chercher les plus courts distances de proche en proche avec des ajustements successifs.
Données : R(X, U, ℓ) (réseau sans circuit absorbant) et s (racine de R)

1. **Initialisation :**
   - S = {s}
   - π(s) = 0
   - π(x) = ∞ pour tout x ∈ X \ {s}
   - x' = s

2. **Tant que** S ≠ X :
   - Pour chaque arc u dont l'extrémité initiale est x' et l'extrémité terminale est y ∉ S :
     - Si π(x') + ℓ(u) < π(y) :
       - π(y) := π(x') + ℓ(u)
     - Choisir un sommet z ∉ S tel que π(z) = min_{y ∉ S} [π(y)] (en cas d'ex-aequos, on procède à un choix arbitraire)
       - x' := z
       - S := S ∪ {y}

3. **Fin de l'algorithme.**
   - Les valeurs π(x) sont les plus courtes distances.


````python
# la condition initiale de l'algorithme
S = ['s']
potentiels = []

successeur_racine = []
for arc in arcs :
  if arc[0] == 's':
    successeur_racine.append(arc[1])

# Initialiser les potentiels des sommets non successeurs de la racine 's' à +inf
for sommet in range(1, nb_sommet):
    if sommet not in successeur_racine:
        potentiels.append([sommet, math.inf])

# Mettre à jour les potentiels pour les successeurs de la racine 's'
for arc in arcs:
    if arc[0] == 's':
        potentiels.append([arc[1], arc[2]])

potentiels = sorted(potentiels, key = lambda x: x[0])
potentiels.insert(0, ['s',0])

print(f"les sommets marquès dans S sont : {S}")
print(f"ensemble de potentiels d'arcs : {potentiels}")
````

La fonction `prochain_voisin` cherche le prochain voisin non marqué ayant le plus petit potentiel dans un graphe représenté par une liste d'arcs et une liste de potentiels. Elle prend en compte les paramètres suivants :

- `S` : Une liste des sommets déjà marqués.
- `potentiels` : Une liste des potentiels des sommets.
- `arcs` : Une liste des arcs du graphe.

La fonction retourne le prochain voisin non marqué avec le plus petit potentiel.

Cette fonction est utilisée dans l'algorithme de Dijkstra pour déterminer le prochain sommet à marquer lors de la recherche du plus court chemin dans un graphe pondéré.


````python
def prochain_pivot(S, potentiels, arcs= arcs, sommets= sommets):
  pivot = 0
  sommet_condidats = []
  sommet_possibles = []
  arc_plus_court_chemin = 0

  for sommet in sommets :
    successeurs_sommet_dans_S = 0
    if sommet in S :
      continue
    else :
      for arc in arcs :
        if arc[0] == sommet and arc[1] in S:
          successeurs_sommet_dans_S += 1
      if successeurs_sommet_dans_S != 0 :
        continue
      else :
        sommet_condidats.append(sommet)

  for sommet in sommet_condidats :
    if potentiels[sommets.index(sommet)][1] not in range(0, 11) :
      continue
    else :
      sommet_possibles.append(sommet)

  if len(sommet_possibles) == 0:
    if sommets[-1] not in S :
      pivot = sommets[-1]
  else:
    pivot = sommet_possibles[0]
    for sommet in sommet_possibles :
      if potentiels[sommets.index(sommet)][1] < potentiels[sommets.index(pivot)][1] :
        pivot = sommet

  for arc in arcs :
    if arc[1] == pivot and potentiels[sommets.index(pivot)][1] == potentiels[sommets.index(arc[0])][1] + arc[2]:
      arc_plus_court_chemin = arc
      break

  return pivot, arc_plus_court_chemin
````

La fonction `mise_a_jour_potentiel` met à jour les potentiels des successeurs du pivot en fonction des arcs du graphe et des potentiels actuels. Elle prend en compte les paramètres suivants :

- `pivot` : Le sommet pivot dont les successeurs doivent être mis à jour.
- `S` : Une liste des sommets déjà marqués.
- `potentiels` : Une liste des potentiels des sommets.
- `arcs` : Une liste des arcs du graphe.

La fonction parcourt les arcs du graphe pour chaque successeur du pivot qui n'est pas déjà marqué. Pour chaque arc, elle calcule le potentiel potentiel_nouveau du successeur en ajoutant la longueur de l'arc au potentiel du pivot. Si ce potentiel potentiel_nouveau est inférieur au potentiel actuel du successeur, elle met à jour le potentiel du successeur avec potentiel_nouveau.

Après avoir mis à jour les potentiels des successeurs du pivot, la fonction ajoute le pivot à la liste des sommets marqués S.

La fonction retourne la liste mise à jour des potentiels et la liste mise à jour des sommets marqués S.

Cette fonction est utilisée dans l'algorithme de Dijkstra pour mettre à jour les potentiels des successeurs du pivot à chaque itération de l'algorithme.


````python
def mise_a_jour_potentiels(pivot, S, potentiels, sommets= sommets, arcs= arcs) :
  successeurs_pivot = []
  for arc in arcs :
    if arc[0] == pivot :
      successeurs_pivot.append(arc)

  for arc in successeurs_pivot :
    if potentiels[sommets.index(arc[1])][1] > (potentiels[sommets.index(pivot)][1] + arc[2]) :
      potentiels[sommets.index(arc[1])][1] = potentiels[sommets.index(pivot)][1] + arc[2]

  S.append(pivot)

  return potentiels, S
````

Le code principal réalise l'algorithme de Dijkstra pour trouver les plus courts chemins dans un graphe pondéré.

1. **Initialisation** :
    - Le compteur d'itération `iteration` est initialisé à 0.
    - L'état initial de `S` (ensemble des sommets marqués) et `potentiels` (ensemble des potentiels d'arcs) est affiché.

2. **Boucle principale** :
    - Tant que tous les sommets ne sont pas marqués :
        - L'itération est incrémentée.
        - La fonction `prochain_pivot` est appelée pour choisir le prochain pivot.
        - Si le pivot est 0, l'algorithme est terminé et un message approprié est affiché.
        - Sinon, les potentiels sont mis à jour à l'aide de la fonction `mise_a_jour_potentiels`.
        - L'état actuel de `S` et `potentiels` est affiché.

3. **Fin de l'algorithme** :
    - Les potentiels finaux et les arcs des plus courts chemins sont affichés.
    - Si le dernier arc est 0, cela indique une terminaison prématurée et il est retiré de la liste.

Ce code fournit une implémentation de l'algorithme de Dijkstra avec des commentaires explicatifs pour chaque étape, facilitant ainsi la compréhension de son fonctionnement.


````python
# Continuer l'algorithme de Dijkstra
iteration = 0  # Initialisation du compteur d'itération
print(f"{iteration} iteration :\nS = {S}\npotentiels = {potentiels}\n")  # Affichage de l'état initial

pc_chemin = []  # Initialisation de la liste des arcs des plus courts chemins
while len(S) < len(sommets):  # Tant que tous les sommets ne sont pas marqués
    iteration += 1  # Incrémentation du compteur d'itération
    # Choisir le prochain pivot parmi les successeurs du pivot actuel qui ne sont pas dans S
    pivot, arc = prochain_pivot(S, potentiels)  # Appel de la fonction prochain_pivot
    pc_chemin.append(arc)  # Ajout de l'arc du chemin le plus court actuel à la liste

    if pivot == 0:  # Si le pivot est 0, l'algorithme est terminé
        if len(S) == nb_sommet - 1:  # Si tous les sommets sauf 's' sont marqués
            print("Terminer! tout les sommets sont marquer : X = S")
        else:  # Sinon, il y a un problème avec 's'
            print("Terminer! on a pas pu marquer tout les sommets : le sommet 's' n'est pas racine")
        break  # Sortie de la boucle
    else:
        # Mettre à jour les potentiels des successeurs du pivot qui ne sont pas dans S
        potentiels, S = mise_a_jour_potentiels(pivot, S, potentiels)  # Appel de la fonction mise_a_jour_potentiels
        print(f"{iteration} iteration :\nS = {S}\npotentiels = {potentiels}\n")  # Affichage de l'état actuel

# Afficher les potentiels finaux
print(f"Potentiels finaux : {potentiels}\nPlus courts chemins : {pc_chemin}")  # Affichage des potentiels finaux et des arcs des plus courts chemins
if pc_chemin[-1] == 0:  # Si le dernier élément de pc_chemin est 0, c'est une terminaison prématurée
    pc_chemin.pop()  # Retrait du dernier élément
````

Le code utilise la bibliothèque NetworkX pour créer et afficher un graphe orienté avec des poids sur les arcs.

1. **Création du graphe** :
    - Un graphe dirigé est initialisé avec `nx.DiGraph()`.
    - Les sommets sont ajoutés au graphe à l'aide de `add_nodes_from(sommets)`.
    - Les arcs pondérés sont ajoutés au graphe avec `add_weighted_edges_from(arcs)`.

2. **Positionnement des sommets** :
    - Les sommets sont positionnés pour l'affichage avec `nx.shell_layout(graphe)`.

3. **Affichage du graphe** :
    - Les sommets et les arcs sont dessinés avec `nx.draw()`.
    - Les labels des arcs avec les poids sont affichés avec `nx.draw_networkx_edge_labels()`.
    - Les arcs des plus courts chemins sont mis en évidence avec `nx.draw_networkx_edges()` en utilisant les données de `pc_chemin`.

4. **Affichage des informations supplémentaires** :
    - Le degré de chaque sommet est imprimé avec une boucle sur les sommets du graphe.
    - Les étiquettes des sommets sont ajoutées au graphe pour un affichage plus clair.
    - Le nombre de sommets et d'arcs est affiché à l'écran.

Ce code permet de visualiser le graphe avec ses sommets, ses arcs pondérés, les labels des arcs et les plus courts chemins mis en évidence. De plus, il affiche des informations supplémentaires telles que le degré de chaque sommet et le nombre de sommets et d'arcs dans le graphe.


````python
graphe = nx.DiGraph()

graphe.add_nodes_from(sommets)
graphe.add_weighted_edges_from(arcs)

# Positionnement des sommets pour l'affichage
pos = nx.shell_layout(graphe)

# Dessiner les sommets et les arcs
nx.draw(graphe, pos, with_labels=True, node_color='SkyBlue', node_size=1000, font_size=15, font_weight='bold')

# Dessiner les labels des arcs avec les poids
edge_labels = nx.get_edge_attributes(graphe, 'weight')
nx.draw_networkx_edge_labels(graphe, pos, edge_labels=edge_labels, font_size=12, font_color='red')

# Mettre en évidence les arcs des plus courts chemins
edges_to_highlight = [(arc[0], arc[1]) for arc in pc_chemin]
nx.draw_networkx_edges(graphe, pos, edgelist=edges_to_highlight, edge_color='tab:green', width=8, alpha= 0.5)

# Afficher le graphe
plt.show()

for sommet in graphe.nodes :
  print(f"degrè de {sommet} = {graphe.degree(sommet)}")
  graphe.nodes[sommet]['label'] = sommet

print(f"nombre de sommets : {graphe.number_of_nodes()}")
print(f"nombre d'arcs : {graphe.number_of_edges()}")
````
